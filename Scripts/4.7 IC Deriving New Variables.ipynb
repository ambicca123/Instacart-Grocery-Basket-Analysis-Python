# If-Statements with User-Defined Functions

```python
# importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# importing dataframe
path = r'/Users/sudip/Desktop/DATA ANALYST/Projects/Project 4 Python/Instacart Basket Analysis'
df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
# creating subset of the dataframe for first one million rows.
df = df_ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# define a function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low range product'
    if (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    if row['prices'] > 15:
        return 'High range'
    else: return 'Not enough Data'
```

```python
#Apply the function
df['price_label'] = df.apply(price_label, axis = 1)
```

# If statements with for-loops

```python
#creating for-loops
result = []
for value in df_ords_prods_merged['orders_day_of_the_week']:
    if value == 0:
        result.append('Busiest Day')
    elif value == 4:
        result.append('Least Busy')
    else :
        result.append('Regularly Busy')
```

```python
result
```

```python
df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# Task 4.7

```python
# creating for-loop for busiest days and slowest days
result2 = []

for value in df_ords_prods_merged['orders_day_of_the_week']:
  if value in [0,1]:
    result2.append("Busiest days")
  elif value in [3,4]:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```

```python
df_ords_prods_merged['busiest_days'] = result2
```

```python
df_ords_prods_merged['busiest_days'].value_counts(dropna= False)
```

The total number of rows for 0 and 1 are 6204182 and 5660230, which is equal to busiest days rows 11864412. The total number of rows for 3 and 4 are 3840534 and 3783802, which is equal to Slowest days rows 7624336. So the values in this new columns are accurate.

```python
#dentifying the busiest hours of the day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating for-loop for sorting the busiest hours of the day by periods
result3 = []
for value in df_ords_prods_merged['order_hour_of_day']:
    if value >=9 and value <=16:
        result3.append('Most orders')
    elif value >=0 and value <=5:
        result3.append('Fewest orders')
    else:
        result3.append('Average orders')
```

```python
# Creating new column 'busiest_period_of_day' with the results of the for-loop
df_ords_prods_merged['busiest_period_of_day'] = result3
```
