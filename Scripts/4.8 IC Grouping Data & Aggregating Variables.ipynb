```python
# importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# importing dataframe
path = r'/Users/sudip/Desktop/DATA ANALYST/Projects/Project 4 Python/Instacart Basket Analysis'
```

```python
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_derived.pkl'))
```

```python
# calculating aggregated mean of the “order_number” column grouped by “department_id”
ords_prods_merge.groupby('department_id')['order_number'].mean()
```

Now, we have the average order details for all the department ids. While calculating the average orders from subset of 1 million rows, there were fewer department ids and mean value was also different from the results with the entire dataframe.

```python
# Creating a loyalty flag for existing customers using the transform() and loc() functions.
# Step one: Create new column with max order
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge.head(10)
```

```python
# Step two: Create loyalty flag column based on max order criteria
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head(5)
```

```python
# checking the basic statistics of the product prices for each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# target different types of spenders 
# Step 1: Create new column of average price paid by users
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(5)
```

```python
# Step 2: Assign spending flag based on average price
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'average_price', 'spending_flag']].head(20)
```

```python
# Create order frequency flag based on median days since previous order.
# Step 1: Create a new column showing median days since previous order.
ords_prods_merge['median_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Step2: Assign frequency flag based on median days since previous order
ords_prods_merge.loc[ords_prods_merge['median_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order'] > 10) & (ords_prods_merge['median_order'] <= 20),'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order'] <= 10,'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'median_order', 'order_frequency_flag']].head(20)
```
